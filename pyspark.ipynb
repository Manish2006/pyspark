PySpark Handling Missing Values

1) Drop Columns

2)Dropping rows

3)Handling Missing values by Mean,Median,Mode **bold text**

```python
df_spark2.show()
```

```python
##Dropping the Null values
df_spark2.na.drop().show(3)
```

```python
##Dropping the Null values where all Row is NULL
df_spark2.na.drop(how='all').show(3)
```

```python
##Threshold it will remove the NAN values where thresh=2 (if 2 rows have NAN values)
df_spark2.na.drop(how='any',thresh=2).show(3)
```

```python
##Subset - it will remove the NAN values based on particlar column .i.e.Row ID
df_spark2.na.drop(how='any',subset=['Row ID']).show(3)
```

```python
##Replace Missing Values with text = Missing Values
df_spark2.na.fill('Missing Values').show(1)
```

```python
##Replace Missing Values with text = Missing Values for Particular One Column
df_spark2.na.fill('Missing Values','Row ID').show(1)
```

```python
## Replacing Null values Mean Value

from pyspark.ml.feature import Imputer
```

```python
imputer=Imputer(
    inputCols=['Shipping Cost','Unit Price','Discount'],
    outputCols=["{}_imputed".format(c) for c in ['Shipping Cost','Unit Price','Discount']]
).setStrategy("mean")
imputer.fit(df_spark2).transform(df_spark2).show(2)
```

```python
##Filtering of data using pyspark

## Discount<=0.01
df_spark2.filter("Discount<=0.01").show()
```

```python
## Discount<=0.01
df_spark2.filter("Discount<=0.01").select(['Row ID','Discount']).show()
```

```python
## other way to filter

df_spark2.filter(df_spark2['Discount']<=0.01).show()
```

```python
##Multiple conditions of filtering
df_spark2.filter((df_spark2['Discount']<=0.01) &
                 (df_spark2['Unit Price']<=2)).show()
```
